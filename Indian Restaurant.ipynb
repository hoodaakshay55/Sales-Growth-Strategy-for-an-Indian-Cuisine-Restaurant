```python

df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')
df['Time'] = pd.to_datetime(df['Time'], format='%H:%M').dt.time
df["Quantity"]=df["Quantity"].astype(int)
df["Price_Per_Unit"]=df["Price_Per_Unit"].astype(int)
df["Total_Amount"]=df["Total_Amount"].astype(int)
df["Customer_Rating"]=df["Customer_Rating"].astype(int)
```

```python
df.info()
```

```python
df.dropna(inplace= True)
```

```python
df.head()
```

```python
df.info()
```

```python
# Create new date-related features
df['Month'] = df['Date'].dt.month_name()
df['Year'] = df['Date'].dt.year
df['Weekday'] = df['Date'].dt.day_name()
```

```python
df.head()
```

```python
# Category Distributions
print(df['Customer_Type'].value_counts())
print(df['Item_Category'].value_counts())
print(df['Payment_Mode'].value_counts())
print(df['Home_Delivery'].value_counts())
```

```python
## 6. Sales & Revenue Analysis
# Total and average revenue
total_revenue = df['Total_Amount'].sum()
avg_order_value = df['Total_Amount'].mean()
print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
```

```python
#Top items and categories
top_items = df.groupby('Item_Name')['Total_Amount'].sum().sort_values(ascending=False).head(10)
top_categories = df.groupby('Item_Category')['Total_Amount'].sum().sort_values(ascending=False)
print(top_items)
print(top_categories)
```

```python
# Revenue by day of week
revenue_by_weekday = df.groupby('Weekday')['Total_Amount'].sum().reindex(
    ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'])

# Revenue by time of day
revenue_by_hour = df.groupby('Time')['Total_Amount'].sum()

# Average Order Value: New vs Returning Customers
aov_by_customer_type = df.groupby('Customer_Type')['Total_Amount'].mean()

print(revenue_by_weekday)
print(revenue_by_hour)
print(aov_by_customer_type)
```

```python
# Ratings
sns.histplot(df['Customer_Rating'], bins=5, kde=False)
plt.title("Distribution of Customer Ratings")
plt.show()
```

```python
# Ratings by category
sns.boxplot(data=df, x='Item_Category', y='Customer_Rating')
plt.xticks(rotation=45)
plt.title("Customer Ratings by Item Category")
plt.show()
```

```python
## 8. Time-based Trends
# Revenue by month
monthly_revenue = df.groupby('Month')['Total_Amount'].sum().sort_values()
monthly_revenue.plot(kind='bar', title='Monthly Revenue')
plt.ylabel('Revenue')
plt.show()
```

```python
# Orders by weekday
orders_by_day = df['Weekday'].value_counts().sort_index()
orders_by_day.plot(kind='bar', title='Orders by Weekday')
plt.ylabel('Number of Orders')
plt.show()
```

```python
# Payment and Delivery
# Payment mode preferences
sns.countplot(data=df, x='Payment_Mode')
plt.title("Payment Mode Distribution")
plt.show()
```

```python
# Home Delivery trends
sns.countplot(data=df, x='Home_Delivery')
plt.title("Home Delivery Orders")
plt.show()
```

```python
## Correlation Heatmap
numerical_cols = df[['Quantity', 'Price_Per_Unit', 'Total_Amount', 'Customer_Rating']]
sns.heatmap(numerical_cols.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```
